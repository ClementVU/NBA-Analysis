```python
# Which position scores the most ?
# We must use the column "POS" that represent the position of the player & take the average by AVG() to answer the problematic 
query = """
    SELECT Pos, AVG(PTS) as AVG_PTS
    FROM `projetnba.nba. datanba` 
    WHERE Pos NOT LIKE '%-%'
    GROUP BY Pos
    ORDER BY  AVG_PTS DESC
    
"""
df_avg_pts_per_pos = client.query(query).to_dataframe()
df_avg_pts_per_pos 
```

```python
# To understand the study, we count the number of players per position
# We delete the duplicate values (player that change team and position during the season 2021-2022)
query = """
    SELECT Pos, COUNT (DISTINCT Player) as `number_of_player`
    FROM `projetnba.nba. datanba`
    WHERE Pos NOT LIKE '%-%' 
    GROUP BY Pos
    
"""
df_players_per_pos = client.query(query).to_dataframe()
df_players_per_pos 
```

```python
# The impact of age on the % Shooting (FG%) in 3pts
# 1st step : Average of 3pts & 2pts of all age players
# 2st step : Remove biaised values (>10 only)

query = """
    SELECT Age, AVG(_3P_) as FG3pts , AVG(_2P_) as FG2pts, AVG(_3P_ + _2P_)/2 as FGtot
    FROM `projetnba.nba. datanba` 
    GROUP BY Age 
    HAVING COUNT(Player) > 10
    ORDER BY FGtot DESC
    
"""
df_impact_age_on_shoot = client.query(query).to_dataframe()
df_impact_age_on_shoot
```
